```python
dev_X,test_X,dev_Y,test_Y=train_test_split(data,labels,test_size=0.20,stratify=labels,random_state=10)
dev_X1,test_X1,dev_Y1,test_Y1=train_test_split(data1,labels,test_size=0.20,stratify=labels,random_state=10)
dev_X2,test_X2,dev_Y2,test_Y2=train_test_split(data2,labels,test_size=0.20,stratify=labels,random_state=10)
```

```python
def imp_stack(test_X,test_X1,test_X2):
    p1=m1.predict(test_X2)
    p1=np.array(p1)
    p2=m2.predict(test_X)
    p2=np.array(p2)
    p3=m3.predict(test_X1)
    p3=np.array(p3)
    p4=m4.predict(test_X)
    p4=np.array(p4)
    p_ans=p1+p2+p3+p4
    # p_ans=p2+p3+p4
    #p_ans=p3+p4
    p_ans=p_ans/4
    p_ans=np.round_(p_ans)
    return p_ans
```
